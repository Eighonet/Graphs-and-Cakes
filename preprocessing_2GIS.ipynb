```python
import json

def validateJSON(jsonData):
    try:
        json.loads(jsonData)
    except ValueError as err:
        return False
    return True

InvalidJsonData = "{""point"":{""lat"":53.6887843,""lon"":91.6732582},""userLocation"":1}"
isValid = validateJSON(InvalidJsonData)

print("Given JSON string is Valid", isValid)

validJsonData = "{""point"":{""lat"":53.6887843,""lon"":91.6732582},""userLocation"":1}"
isValid = validateJSON(validJsonData)

print("Given JSON string is Valid", isValid)
```

```python
import json
import jsonschema
from jsonschema import validate

# Describe what kind of json you expect.
studentSchema = {
    "point": {
        "lat": {"type": "number"},
        "lon": {"type": "number"},
    },
    "userLocation": "number",
}

def validateJson(jsonData):
    try:
        validate(instance=jsonData, schema=studentSchema)
    except jsonschema.exceptions.ValidationError as err:
        return False
    return True

# Convert json to python object.
jsonData = json.loads(routes.iloc[0, 15])
# validate it
isValid = validateJson(jsonData)
if isValid:
    print(jsonData)
    print("Given JSON data is Valid")
else:
    print(jsonData)
    print("Given JSON data is InValid")

# Convert json to python object.
jsonData = json.loads(routes.iloc[0, 2])
# validate it
isValid = validateJson(jsonData)
if isValid:
    print(jsonData)
    print("Given JSON data is Valid")
else:
    print(jsonData)
    print("Given JSON data is InValid")
```
